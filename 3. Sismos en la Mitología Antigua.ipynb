## 1. Introducción <a id="1-introducción"></a>

Desde sus orígenes, el ser humano ha buscado explicaciones para los fenómenos naturales que lo rodean, especialmente aquellos que, por su fuerza y repentinidad, despertaban temor e incertidumbre. Los terremotos, con su capacidad de transformar el paisaje en instantes, fueron interpretados por distintas civilizaciones a través de mitos y relatos que reflejaban su cosmovisión, creencias religiosas y relación con la naturaleza. Estas narrativas no solo ofrecían una explicación, sino que también influían en rituales, construcciones y respuestas ante desastres.

Con el paso del tiempo y el desarrollo del conocimiento científico, se desplazaron estas interpretaciones míticas, dando paso a la sismología moderna.

## 2. Sismología en la Antigüedad <a id="2-sismología-en-la-antigüedad"></a>

### 2.1 Mitología Griega <a id="21-mitología-griega"></a>
Según la mitología griega, Poseidón, dios del mar y de los movimientos terrestres, provocaba alteraciones en la Tierra al enfurecerse y golpearla con su tridente. Además, diversas ciudades como Micenas, Éfeso, Cnido y Hierápolis fueron edificadas sobre fallas geológicas, con estructuras sagradas ubicadas estratégicamente de manera que la falla las atravesara.

Algunas posibles explicaciones para esta elección incluyen las "barreras naturales" que proporcionaban estas zonas, así como la presencia de fuentes termales y cuerpos de agua subterráneos, elementos de gran relevancia en la cultura griega.

Antes de Aristóteles, se pensaba que los continentes flotaban sobre la superficie terrestre y que su desplazamiento era la causa de los terremotos. Otra hipótesis sugería que la Tierra experimentaba movimientos debido a la expansión y contracción generadas por la humedad o la sequedad extrema.

No fue sino hasta aproximadamente el año 300 a. C. que Aristóteles formuló su teoría, en la que sostenía que los terremotos eran provocados por corrientes de aire atrapadas en el interior de la Tierra (“exhalaciones”).

### 2.2 Mitología Nórdica <a id="22-mitología-nórdica"></a>
El poema Völuspá, parte de la Edda poética, relata terremotos y otras calamidades en el contexto del fin del mundo, denominado Ragnarök.

En esta obra, los terremotos son presentados como señales inequívocas que marcan el inicio de Ragnarök. El poema describe cómo el mundo se desintegrará a través de escenas de devastación, donde la tierra tiembla, el cielo colapsa y el mar inunda las tierras. Para los vikingos, estos sismos eran vistas como advertencias de la intervención de fuerzas sobrenaturales que alteraban el equilibrio del mundo.

Uno de los seres vinculados a los terremotos es Loki, el dios travieso y desordenado. Tras ser responsable de la muerte del dios Baldr, Loki fue castigado por los demás dioses, quienes lo ataron y le colocaron una serpiente sobre su cabeza, cuyo veneno caía sobre él. Cada vez que el veneno lo tocaba, Loki se retorcía de dolor, provocando temblores en la tierra.

Un personaje esencial para comprender los terremotos en la mitología vikinga es Jörmungandr, la gigantesca serpiente que rodea el mundo. Según la Völuspá, cuando llegue el fin del mundo, esta criatura saldrá del mar, desatando caos, olas gigantescas y terremotos.

El poema también vincula los terremotos a otras entidades poderosas como los gigantes (jötnar) y el lobo Garmr. Los vikingos creían que los terremotos eran señales de que las fuerzas caóticas se liberaban de sus prisiones y que la batalla final entre dioses y monstruos estaba por comenzar.

### 2.3 Cultura China <a id="23-cultura-china"></a>
Durante la dinastía Han, comenzó el registro organizado de los terremotos. Un hito importante en este proceso fue la invención del primer sismoscopio en el año 132 d.C. por Zhang Heng, un dispositivo innovador que marcó un gran avance en la detección de sismos y estableció las bases para futuras investigaciones científicas. Además, China desarrolló explicaciones naturalistas tempranas, asociando los terremotos con bloqueos del "qi", lo cual refleja una visión avanzada para su tiempo.

En la mitología, figuras como Fuxi y Nüwa, representadas con rostros humanos y cuerpos de serpiente, se vinculan con la creación y restauración del cielo, un evento considerado el primer terremoto de la humanidad. Otro mito importante es el de Zhu Long, un dragón que vivía debajo de la montaña Zhong, cuya respiración, según las creencias, causaba terremotos al desplazarse bajo la Tierra, asociándose con el poder de las aguas y los ríos.

Más allá de las leyendas, filósofos como Zhuangzi propusieron teorías como la de la Tierra flotante, que explicaba los sismos como el resultado del choque de aguas subterráneas, mientras que la teoría del Yin y Yang interpretaba los terremotos como un desequilibrio entre estas fuerzas opuestas. Esta última teoría fue promovida por Bo Yangfu, quien asociaba los sismos con eventos sociales, como la sequía de ríos y el colapso de dinastías.

Durante las dinastías Han y Qi, se registraron más de 120 terremotos, lo que permitió el desarrollo de protocolos para enfrentarlos y dio origen a la fenomenología sísmica, una disciplina que intentaba correlacionar los sismos con fenómenos astronómicos, como eclipses y lluvias de meteoros, aunque todavía perduraban ideas mágicas y especulativas. Zhang Heng, con su sismoscopio, fue capaz de registrar con precisión el terremoto de Longxi en el 134 d.C., un avance significativo en el campo de la sismología.

### 2.4 Cultura Japonesa <a id="24-cultura-japonesa"></a>
En la literatura clásica japonesa, los terremotos se mencionaban muy poco, ya que las técnicas de construcción empleadas en esa época otorgaban gran resistencia a las estructuras. De manera similar, al erigir las columnas, se colocaban sobre pedestales hechos de guijarros, lo que permitía que la estructura tuviera un movimiento independiente del suelo, ayudando a dispersar la energía de los terremotos.

Otra razón por la que los registros de sismos en Japón son escasos está relacionada con la visión que tenían los japoneses sobre estos eventos. Para ellos, los terremotos eran considerados un castigo para los gobernantes que oprimían a la población, por lo que documentar estos eventos o sus consecuencias era percibido como una falta de respeto, lo que resultaba en castigos para los autores.

Los fenómenos sísmicos eran atribuidos a los movimientos de deidades o criaturas que sostenían las principales islas de Japón, como dioses, gigantes, bueyes, dragones, serpientes y peces.

Durante el siglo XVIII, un dios en forma de pez o siluro, llamado Onamazu o Namazu, reemplazó a estas deidades. El terremoto ocurrido en el periodo Edo (siglo XVIII) en 1855 fue interpretado como un castigo de Namazu debido a la avaricia humana, lo que obligó a la redistribución de riquezas entre la población.

Para los japoneses, Namazu no era un dios ni positivo ni negativo, sino un ser encargado de la rectificación del mundo. Había sido descrito como un dios ubicado en las profundidades del mar, y se creía que al mover su cola causaba movimientos sísmicos y tsunamis que afectaban Japón. Esta creencia surgió de la observación de que los siluros mostraban comportamientos inusuales antes de la ocurrencia de un terremoto.

### 2.5 Cultura Mexica <a id="25-cultura-mexica"></a>
Los mexicas percibían el universo como un ciclo continuo de creación y destrucción, con diversas eras alternadas por catástrofes. En su cosmología, los terremotos eran vistos como el resultado del choque entre la Tierra y el Sol. Esta visión ilustra la estrecha relación entre los fenómenos naturales y las creencias religiosas, considerando los sismos no solo como sucesos físicos, sino también como expresiones cósmicas. En el periodo prehispánico, se documentaron al menos 22 terremotos, siete de los cuales alcanzaron una magnitud de 6 o más grados. 

Un mito destacado relacionado con los sismos es el que involucra a Tepeyollotl, el dios de los temblores, las cuevas y las montañas. Para los mexicas, los temblores eran causados por este dios, ya sea como castigo por la falta de sacrificios o como un mal necesario para preservar el equilibrio del entorno natural.

## 3. Similitudes entre Culturas <a id="3-similitudes-entre-culturas"></a>

A pesar de las diferencias geográficas, geológicas y culturales, varias civilizaciones compartían similitudes en su interpretación de los terremotos:
1. Uso de animales zoomorfos: Las culturas costeras (Grecia, Nórdica, Japón) vinculaban los terremotos con criaturas acuáticas, como serpientes o peces gigantes, mientras que las culturas continentales (China, Mexica) los asociaban a seres terrestres como dragones y jaguares.

2. Connotación del terremoto: La mayoría veía los terremotos como castigos o ajustes cósmicos (Griegos, Nórdicos, Japoneses, Mexicas), mientras que China los interpretaba como desequilibrios naturales.

3. Transición a explicaciones físicas: Grecia y China avanzaron más rápido hacia explicaciones racionales (como las teorías de Aristóteles o el sismoscopio de Zhang Heng), mientras que Japón mantuvo mitos pero desarrolló técnicas antisísmicas prácticas.

4. Conexión con ciclos cósmicos: Las culturas apocalípticas (Nórdicos, Mexicas, Japoneses) veían los terremotos como parte de ciclos cósmicos de destrucción y renacimiento.

5. Adaptación práctica: Las culturas con mayor exposición a terremotos desarrollaron soluciones prácticas, como la arquitectura antisísmica en Japón y los protocolos de registro en China.

## 4. Conclusiones <a id="4-conclusiones"></a>
Los mitos sísmicos deben entenderse no como simples relatos ficticios, sino como respuestas culturales ante la vulnerabilidad humana frente a los fenómenos naturales. Cada civilización desarrolló sus interpretaciones de los terremotos en función de su contexto geológico (como islas volcánicas o zonas con fallas geológicas activas) y su cosmovisión religiosa, que podría incluir dioses vengativos o la noción de ciclos cósmicos. Estas narrativas, aunque actualmente reemplazadas por explicaciones científicas, reflejan el esfuerzo humano por comprender lo desconocido y por imponer un sentido de orden ante la imprevisibilidad y el caos de los desastres naturales.
