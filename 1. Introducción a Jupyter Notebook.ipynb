## ¿Qué es Jupyter Notebook?<a id="1-Introducción"></a>
Es una aplicación web para crear y compartir documentos computacionales que integran cálculos, resultados, texto explicativo, expresiones matemáticas, imágenes y representaciones multimedia.

## Ejecutar celdas <a id="2-Ejecutar-celdas"></a>
El componente esencial de un cuaderno Jupyter es la celda. Primero vamos a aprender a ejecutar celdas:
- `Shift + Enter` → Ejecuta la celda y pasa a la siguiente.  
- `Ctrl + Enter` → Ejecuta la celda y permanece en la misma.  

```python
#Ejecuta la celda
x = 3*2
print(x)
```

### Celda de código

Las celdas de código presentan tres áreas principales: **entrada, visualización y salida**. 

El área de entrada se identifica por el indicador `In []`, ubicado a la izquierda. Dentro de los corchetes puede aparecer:
- un número: refleja el orden de ejecución
> `In [1]`→ corresponde a la primera celda ejecutada tras abrir el cuaderno
- un espacio en blanco: indica que la celda aún no se ha ejecutado
> `In [ ]`
- un asterisco: señala que la celda se encuentra en ejecución 
> `In [*]` 

Sin embargo, si aparece un número, este puede corresponder a ejecuciones previas, lo que implica que la salida no siempre refleja resultados actualizados.

```python
# Esta es una celda de código sin ejecutar
```
