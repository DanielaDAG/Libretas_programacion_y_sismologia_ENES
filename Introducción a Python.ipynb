## 2. Operadores de asignación <a id="2-asignación"></a>
Los operadores de asignación se usan para asignar valores a variables.
**Operador Básico (`=`)**
Asigna un valor a una variable.

```python
x = 10
print("x =", x) 
```

**Operadores Combinados**

 | Operador | Ejemplo | Equivalente | Descripción |
 |----------|---------|-------------|-------------|
 | `+=`     | `x += 2`| `x = x + 2` | Suma y asigna |
 | `-=`     | `x -= 3`| `x = x - 3` | Resta y asigna |
 | `*=`     | `x *= 4`| `x = x * 4` | Multiplica y asigna |
 | `/=`     | `x /= 2`| `x = x / 2` | Divide y asigna |
 | `//=`    | `x //=3`| `x = x //3` | División entera y asigna |
 | `%=`     | `x %= 3`| `x = x % 3` | Módulo y asigna |
 | `**=`    | `x **=2`| `x = x **2` | Potencia y asigna |

```python
a = 5
a += 3  # Equivalente: a = a + 3
print("a += 3 →", a)  

b = 10
b /= 2  # Equivalente: b = b / 2
print("b /= 2 →", b) 
```

Python permite asignar múltiples variables en una sola línea:

```python
x, y, z = 1, 2, 3
print(f"x={x}, y={y}, z={z}")
```

## 3. Operadores aritméticos <a id="3-aritméticos"></a>

| Operador | Símbolo | Ejemplo   | Resultado |

## 4. Operadores lógicos <a id="4-bool"></a>
Los operadores lógicos se utilizan para combinar expresiones condicionales y evaluar valores booleanos. Un valor booleano es un tipo que sólo puede tomar valores `True` o `False`. 
| Operador | Nombre       | Ejemplo           | Descripción                                      |
|----------|--------------|-------------------|--------------------------------------------------|
| `and`    | AND lógico   | `x > 5 and x < 10`| Verdadero si AMBAS condiciones son verdaderas    |
| `or`     | OR lógico    | `x < 5 or x > 10` | Verdadero si AL MENOS UNA condición es verdadera |
| `not`    | NOT lógico   | `not(x < 5)`      | Invierte el resultado: True → False, False → True |

```python
a = True
b = False

print("AND:", a and b)  
print("OR:", a or b)    
print("NOT:", not a)
```

### Tablas de verdad

#### AND (`and`)
| A     | B     | A and B |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | False   |
| False | True  | False   |
| False | False | False   |

#### OR (`or`)
| A     | B     | A or B  |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | True    |
| False | True  | True    |
| False | False | False   |

#### NOT (`not`)
| A     | not A |
|-------|-------|
| True  | False |
| False | True  |

Python evalúa los operadores lógicos de izquierda a derecha y se detiene tan pronto como se determina el resultado final:

- Para `and`: Si el primer operando es `False`, el segundo no se evalúa
- Para `or`: Si el primer operando es `True`, el segundo no se evalúa

```python
# Verificación de rangos
edad = 25
print("¿Es adulto joven?", edad >= 18 and edad <= 30)  

# Validación de entrada
usuario_valido = True
contrasena_valida = False
print("¿Acceso permitido?", usuario_valido and contrasena_valida)  

# Condición negada
llueve = True
print("¿No llueve?", not llueve)

#Evaluación de operadores
psicometricos = False
medicos = True
print("¿Candidato válido?", psicometricos and medicos)
```

#### Operadores con Valores no Booleanos

Python considera como `False`:
- `None`
- `False`
- `0` (cualquier tipo numérico)
- Secuencias vacías (`""`, `[]`, `()`)
- Diccionarios vacíos `{}`

Todos los demás valores se consideran `True`.

```python
print("'Hola' and 'Mundo' =", "Hola" and "Mundo")  
print("0 or 100 =", 0 or 100)                      
print("not [] =", not [])  
```

## 8. Listas <a id="8-listas"></a>
Las listas en Python son colecciones ordenadas de objetos. Pueden contener cualquier tipo de dato y son mutables, lo que significa que puedes cambiar su contenido.

```python
# Creación de una lista
osos = ['pandas', 'polares', 'pardos']
```

**Indexación y modificación:**
<br>
Las listas están ordenadas y se pueden acceder mediante índices. También se pueden modificar.

```python
# Acceso a elementos de la lista
print(osos[0])  # 'pandas'
print(osos[0:2])  # ['pandas', 'polares']
print(osos[-1])  # 'pardos'

# Modificación de la lista
osos.append('rojo')
print(osos)
osos[1] = 'azules'
print(osos)
```

**Iteración:**
<br>
Puedes iterar a través de una lista usando un bucle `for`.

```python
# Iteración a través de una lista
for oso in osos:
    print(oso)
```

## 11. Funciones <a id="11-funciones"></a>
Las funciones en Python se definen usando la palabra clave `def`. Aquí hay un ejemplo de una función simple:

```python
# Definición de una función
def al_cuadrado(n):
    """
    Devuelve una lista de potencias de 2, 
    comenzando desde 0 y progresando hasta
    que se generen n elementos.
    """
    resultados = []
    
    for i in range(n):
        resultados.append(2**i)
        
    return resultados
```

```python
# Llamada a la función
al_cuadrado(10)
```

```python
def regre(x,y):
    #Promedios
    xmean=np.mean(x)
    ymean=np.mean(y)  
    #Errores 
    
    #Errores cuadráticos
    Sxx=sum((x-xmean)**2)
    Sxy=sum((x-xmean)*(y-ymean))
    Syy=sum((y-ymean)**2)
    
    m=Sxy/Sxx #Obtención de pendiente
    b=(sum(y)-m*sum(x))/len(x) #Ordenada al origen 
    rxy=Sxy/((Sxx*Syy)**0.5) #R
    
    #Error cuadrático medio
    yest=b+m*x #Valores estimados
    err=y-yest
    rms=(sum(err**2)/len(x))**0.5
    tst = rxy/(np.sqrt((1-rxy**2)/(len(x)-2)))

    return(m,b,rxy,err,rms,tst)
```

```python
def Reglin(x,y):
    
    #Promedios
    xmean=np.mean(x)
    ymean=np.mean(y)  
    #Errores 
    
    #Errores cuadráticos
    Sxx=sum((x-xmean)**2)
    Sxy=sum((x-xmean)*(y-ymean))
    Syy=sum((y-ymean)**2)
    
    B1=Sxy/Sxx #Obtención de pendiente
    Bo=(sum(y)-B1*sum(x))/len(x) #Ordenada al origen 
    
    rxy=Sxy/((Sxx*Syy)**0.5) #Coeficiente de Pearson
    
    #Error cuadrático medio
    yest=Bo+B1*x #Valores estimados
    err=y-yest
    rms=(sum(err**2)/len(x))**0.5
    
    #if fig==True:
    #    Regplot(x,y,yest,a=1)

    print(B1,Bo,rxy,rms)
    return B1,Bo,rxy,yest,err,rms


def Regplot(x,y,yest=1.,a=0, X="Eje x",Y="Eje y"):   
    
    if a==1:
        
        fig=plt.figure(figsize=(15,10))
        plt.title("REGRESIÓN LINEAL")
        plt.plot(x,yest, c='r',label="Datos estimados")
        plt.scatter(x,y, c="g",label="Datos reales", marker='*')
    
        plt.xlabel(X)# Titulo de x
        plt.ylabel(Y) #Titulo de y
        #a= "rms="+" " + str(rms)
        b="y="+str(B1)+ " "+"x"+" "+"+"+ " "+str(Bo)
        plt.legend(loc="best")
        #plt.text(max(x)-0.5*max(x),min(y),a,bbox=dict(facecolor='b', alpha=0.3))
        plt.text(max(x)-0.5*max(x),min(y)+0.05*max(y),b,bbox=dict(facecolor='b', alpha=0.3))
    if a==0:
        B1,Bo,rxy,yest,err,rms=Reglin(x,y) 
        
        fig=plt.figure(figsize=(15,10))
        plt.title("REGRESIÓN LINEAL")
        plt.plot(x,yest, c='r',label="Datos estimados")
        plt.scatter(x,y, c="g",label="Datos reales", marker='*')
    
        plt.xlabel(X)# Titulo de x
        plt.ylabel(Y) #Titulo de y
        a= "rms="+" " + str(rms)
        b="y="+str(B1)+ " "+"x"+" "+"+"+ " "+str(Bo)
        plt.legend(loc="best")
        plt.text(max(x)-0.5*max(x),min(y),a,bbox=dict(facecolor='b', alpha=0.3))
        plt.text(max(x)-0.5*max(x),min(y)+0.05*max(y),b,bbox=dict(facecolor='b', alpha=0.3))

        
def Regplot2(x,y, X="Eje x",Y="Eje y"):
        
    fig=plt.figure(figsize=(15,10))
    plt.title("REGRESIÓN LINEAL")
    plt.plot(x,yest, c='r',label="Datos estimados")
    plt.scatter(x,y, c="g",label="Datos reales", marker='*')

    plt.xlabel(X)# Titulo de x
    plt.ylabel(Y) #Titulo de y
    a= "rms="+" " + str(rms)
    b="y="+str(B1)+ " "+"x"+" "+"+"+ " "+str(Bo)
    plt.legend(loc="best")
    plt.text(max(x)-0.5*max(x),min(y),a,bbox=dict(facecolor='b', alpha=0.3))
    plt.text(max(x)-0.5*max(x),min(y)+0.05*max(y),b,bbox=dict(facecolor='b', alpha=0.3))
```

```python
frase = "¿Pvsenqb Pnrfne? Cersvreb rafnynqn Pnrfne"

clave = {'a':'n', 'b':'o', 'c':'p', 'd':'q', 'e':'r', 'f':'s', 'g':'t', 'h':'u',
       'i':'v', 'j':'w', 'k':'x', 'l':'y', 'm':'z', 'n':'a', 'o':'b', 'p':'c',
       'q':'d', 'r':'e', 's':'f', 't':'g', 'u':'h', 'v':'i', 'w':'j', 'x':'k',
       'y':'l', 'z':'m', 'A':'N', 'B':'O', 'C':'P', 'D':'Q', 'E':'R', 'F':'S',
       'G':'T', 'H':'U', 'I':'V', 'J':'W', 'K':'X', 'L':'Y', 'M':'Z', 'N':'A',
       'O':'B', 'P':'C', 'Q':'D', 'R':'E', 'S':'F', 'T':'G', 'U':'H', 'V':'I',
       'W':'J', 'X':'K', 'Y':'L', 'Z':'M'}

resultado = ""
for letra in frase:
    if letra not in clave:
        resultado += letra
    else:
        resultado += clave[letra]
print(resultado)
```

Escribe un programa en Python que genere todas las estrofas de la canción "10 elefantes se columpiaban".

```python
def elefantes_se_columpian(numero):
    return f"{numero} elefantes se columpiaban sobre la tela de una araña, como veían que resistía, fueron a llamar a otro elefante."

# Generar la canción desde 1 hasta 10 elefantes
for i in range(1, 11):
    if i == 1:
        print("1 elefante se columpiaba sobre la tela de una araña, como veía que resistía, fue a llamar a otro elefante.")
    else:
        print(elefantes_se_columpian(i))
```
